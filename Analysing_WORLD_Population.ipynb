## Q3. Growth plot of all the country. Does any country have decrement in any year?

```python
dec=[]
inc=[]
for i in range(len(population)):
    if population["1970 Population"][i]<=population["1980 Population"][i] and population["1980 Population"][i]<=population["1990 Population"][i] and population["1990 Population"][i]<=population["2000 Population"][i] and population["2000 Population"][i]<=population["2010 Population"][i] and population["2010 Population"][i]<=population["2015 Population"][i] and population["2015 Population"][i]<=population["2020 Population"][i] and population["2020 Population"][i]<=population["2022 Population"][i]:
        a=population["Country"][i]
        inc.append(a)
    else:
        b=population["Country"][i]
        dec.append(b)

pd.DataFrame({"Country":dec})
```

## Q4. Countries With 1% or More of the World's Population.

```python
country=[]
for i in range(len(population)):
    if population["World Population Percentage"][i]>=1:
        a=population["Country"][i]
        country.append(a)
pd.DataFrame({"Country":country})
```

## Q5. Condition of country before covid and after covid.

```python
#creating new column "Avg_population_before_COVID".
population["Avg_population_before_COVID"]=round(((population["1970 Population"] + population["1980 Population"] + population["1990 Population"] + population["2000 Population"] + population["2015 Population"])/6),0)
```

```python
#creating new column "Avg_population_after_COVID".
population["Avg_population_after_COVID"]=round(((population["2022 Population"] + population["2020 Population"])/2),0)
```

## Q.10. Which country's population increased and Which country's population decreased after COVID ?

```python
dec_pop=[]
inc_pop=[]
for  i in range(len(population)):
    if population["Avg_population_before_COVID"][i]>=population["Avg_population_after_COVID"][i]:
        DEC=population["Country"][i]
        dec_pop.append(DEC)
    else:
        INC=population["Country"][i]
        inc_pop.append(INC)
        
print("Country's Population decreases after Covid: \n",dec_pop,"\n")
print("Country's Population increases after Covid: \n",inc_pop,"\n")
```
